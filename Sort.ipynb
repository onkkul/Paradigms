```python
import random

class Sort:
    """ All sorting algorithms. """
    def __init__(self):
        return

    
    def bubbleSort(self, array):
        """ Works By: Repeatedly swapping the adjacent elements if they are in wrong order
        Time: Best = O(n), Avg = Worst = O(n^2)
        Space: O(1)
        
        """
        for i in range(len(array)):
            for j in range((len(array)-i-1)):
                if array[j] > array[j+1]:
                    array[j], array[j+1] = array[j+1], array[j]
        return array
    
    
    def selectionSort(self, array):
        """
        Works By: Finding min of unsorted array and placing it in sorted one.
        Time: Best = Avg = Worst = O(n^2)
        Space: O(1)
        
        """
        for i in range(len(array)):
            minimum  = array[i]
            index_of_min = i
            for j in range(i+1, len(array)):
                if minimum > array[j]:
                    minimum = array[j]
                    index_of_min = j
            array[i], array[index_of_min] = array[index_of_min], array[i]
        return array
        
        
    def insertionSort(self, array):
        """
        Works By: Placing individual element in it's correct place like sorting cards
        Time: Best = O(n), Avg = Worst = O(n^2)
        Space: O(1)
        
        """
        for j in range(1, len(array)):
            key = array[j]
            i = j - 1
            while i >= 0 and array[i] > key:
                array[i+1] = array[i]
                i -= 1
            array[i+1] = key
        return array
    
    
    def merge(self, array, left, right):
        """ Helper function for mergeSort. """
        left.append(float("inf"))
        right.append(float("inf"))
        
        i = j = k = 0
        for k in range(len(array)):
            if left[i] < right[j]: 
                array[k] = left[i] 
                i+= 1
            else:
                array[k] = right[j] 
                j+= 1  
        return
    
    def mergeSort(self, array):
        """
        Works By: splitting array into two arrays and recursively sorting the two splitted arrays
        Time: Best = Avg = Worst = O(nlog(n))
        Space: O()
        
        """
        if len(array) > 1:
            middle = len(array)//2
            left, right = array[:middle], array[middle:]
            
            self.mergeSort(left)
            self.mergeSort(right)

            self.merge(array, left, right)
        return array
    
    
    def partition(self, array, start, end):
        """ Helper function for quicksort. 
        returns index of element that is in it's correct place. """
        randomChoice = random.randrange(start, end)
        array[end], array[randomChoice] = array[randomChoice], array[end]
        pivot = array[end]
        
        i = start-1
        for j in range(start, end):
            if array[j] < pivot:
                i += 1
                array[i], array[j] = array[j], array[i]
        
        array[i+1], array[end] = array[end], array[i+1]
        return i+1
    
    def quickSort(self, array, start, end):
        """
        Works By: Splitting elements of array such that elements 
        less than pivot are in one array and greater than pivot are in other.
        Time: Best = Avg = O(nlog(n)), Worst = O(n^2)
        Space: O()
        
        """
        if start < end:
            pivot = self.partition(array, start, end)
            self.quickSort(array, start, pivot -1)
            self.quickSort(array, pivot+1, end)
        return array
    
    
    def heapify(self, array, heap_size, index):
        """ Helper fun for heapSort. forms minHeap of given array. """
        largest = index
        left = 2*index + 1
        right = 2*index + 2
        
        if left < heap_size and array[index] < array[left]:
            largest = left
        if right < heap_size and array[largest] < array[right]:
            largest = right
        if largest != index:
            array[index], array[largest] = array[largest], array[index]
            self.heapify(array, heap_size, largest)
        return
    
    def heapSort(self, array):
        """
        Works By: Extracting min of array using minHeap.
        Time: Best = Avg = Worst = O(nlog(n))
        Space: O()
        
        """
        heap_size = len(array)
        for index in range(heap_size//2 -1, -1, -1):
            self.heapify(array, heap_size, index)
        
        for index in range(heap_size-1, 0, -1):
            array[index], array[0] = array[0], array[index]
            self.heapify(array, index, 0)

        return array
    
    
    def countingSort(self, array):
        """
        Works By: Hashing each element and counting number of occurences
        of each element
        Time: Best = Avg = Worst = O(n+k)
        Space: O()
        
        """
        max_val = float("-inf")
        for element in array:
            max_val = max(max_val, element)

        size = max_val + 1
        count = [0] * size             

        for element in array:
            count[element] += 1             
        
        index = 0
        for element in range(size):            
            for each_count in range(count[element]):  
                array[index] = element
                index += 1
        return array
    
    def radixSort(self,):
        """
        Works By:
        Time: Best = Avg = Worst = O(d(n+k))
        Space: O()
        
        """
        return
    
    def BucketSort(self,):
        """
        Works By:
        Time: Best = Avg = O(n+k), Worst = O(n^2)
        Space: O()
        
        """
        return
    
```
